```python
import numpy as np
import math
```

```python
def printNode(space, count, memLv, A):
  print('\t'*space, sep = '', end = '')
  for member in range(count, (count + memLv) if count + memLv <= A.shape[0] else A.shape[0]):
    print(A[member],'\t', '\t'*((space*2)+1),sep = '', end = '')
  print('\n')
  return

def printPath(space, count, memLv, sizeA):
  print('\t'*space, sep = '', end = '')
  cond = range(count, (count + memLv) if count + memLv <= sizeA else sizeA)
  
  for member in cond:
    if member == cond[-1]:
      print('|\t',sep = '', end = '')
    else:
      print('|\t', '\t'*((space*2)+1),sep = '', end = '')
  print('\n')

def printLine(space, count, memLv, sizeA):
  print('\t'*space, sep = '', end = '')
  cond = range(count, (count + memLv) if count + memLv <= sizeA else sizeA)
  c = 0
  for member in cond:
    if(c%2 == 0 and member == sizeA - 1):
      print(' ____'*((space*2)),sep = '', end = '')
    elif(c%2 == 0):
      print(' ____'*((space*2)+2),sep = '', end = '')
    else:
      print('\t'*((space*2)+2),sep = '', end = '')
    c += 1
  print('\n')

def printTree(A, i = 0, count = 0):
  x = A.shape[0]
  n = int(math.log2(x+1)) if math.log2(x+1)%1 == 0 else int(math.log2(x+1)+1)
  if i <= n-1:
    lv_member = 2 ** i
    space = (2 ** (n - i)) - 1
    if(i > 0):
      printLine(space, count, lv_member, A.shape[0])
      printPath(space, count, lv_member, A.shape[0])
    printNode(space, count, lv_member, A)
    if(i < n - 1):
      printPath(space, count, lv_member, A.shape[0])
    count = count + (2 ** i)
    printTree(A, i+1, count)
```

```python
def parent(i):
  return i // 2

def left(i):
  return 2 * i

def right(i):
  return (2 * i) + 1

def maxHeapify(A, i):
  l = left(i)
  r = right(i)
  
  if l <= A.shape[0] - 1 and A[l] > A[i]:
    largest = l
  else:
    largest = i
  
  if r <= A.shape[0] - 1 and A[r] > A[largest]:
    largest = r
  
  if largest != i:
    A[i], A[largest] = A[largest], A[i]
    maxHeapify(A, largest)

def build_maxHeap(A):
  A_heapSize = A.shape[0]
  for i in range(parent(A.shape[0]), -1, -1):
    maxHeapify(A, i)
  
def Heapsort(A):
  build_maxHeap(A)
  for i in range(A.shape[0] - 1, -1, -1):
    A[0], A[i] = A[i], A[0]
    build_maxHeap(A[:i])
    printTree(A)
  
Arr_test = np.random.randint(0, 30, size = 10)
print(Arr_test)
Heapsort(Arr_test)
print(Arr_test)
```

when n%1 not equal to 0 then n = n+1 <br><br>

Space before number of level is<br>
$2^{n-i}-1$<br>
when<br>
i = index of level(start from 0 to n-1)
