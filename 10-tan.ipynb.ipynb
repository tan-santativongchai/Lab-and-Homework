```python
from collections import Counter

#every node object will have two children, otherwise is a leave
class Node(object):
    def __init__(self, left=None, right=None):
        self.left = left
        self.right = right
    
    def getChild(self):
        return self.left, self.right

def get_code(node, code = ''):
    
    if type(node) is str:
        #stop!!!
        return {node : code}
    
    #get the children
    left, right = node.getChild()
    
    #recursive function
    huffman_code = dict()
    huffman_code.update(get_code(left, code+'0'))
    huffman_code.update(get_code(right, code+'1'))
    
    return huffman_code

def decode(message, huffman_code):
    msge = list(message)
    ori_msg = list()
    while msge != []:
        msg = str()
        while msg not in huffman_code.values() and msge != []:
            msg = msg + msge.pop(0)
            if (len(msge) == 0 or len(msg) == len(max(list(huffman_code.values()), key=len))) and msg not in huffman_code.values():
                return "Error, there is no original message for this huffman code."
        ori_msg.append(list(huffman_code.keys())[list(huffman_code.values()).index(msg)])
    return ''.join(ori_msg)

def calculateTotalCost(freq_dict, huffman_code):

    header = ['Char', 'Count', 'Code', 'msg bit', 'ref table']
    total_bit = 0
    total_tab = 0
    print('{:<15s}{:<15s}{:<15s}{:<15s}{:<15s}'.format(header[0], header[1], header[2], header[3], header[4]))
    l = len(max(huffman_code.values(), key = len))
    for i in sorted(huffman_code.keys()):
        print(f'{i:<15s}{freq_dict[i]:<15.0f}{huffman_code[i]:<15s}{freq_dict[i]*len(huffman_code[i]):<15.0f}{len(huffman_code[i]) + (2 ** l):<15.0f}')
        total_bit = total_bit + freq_dict[i]*len(huffman_code[i])
        total_tab = total_tab + len(huffman_code[i]) + (2 ** l)
    print('{:>45s}{:<15.0f}{:<15.0f}'.format('Total ', total_bit, total_tab))

def make_the_tree(freqs_sorted):
    
    #as long as freqs_sorted.length > 1
    while len(freqs_sorted) > 1:
        
        #combine the two smallest one
        key1, value1 =  freqs_sorted[0]
        key2, value2 =  freqs_sorted[1]
        
        #delete them
        freqs_sorted = freqs_sorted[2:]
        
        #add the new combination to freqs_sorted
        new_value = value1 + value2
        new_node  = Node(key1, key2)
        
        #add to freqs_sorted
        freqs_sorted.append((new_node, new_value))
                
        #sort again!!
        freqs_sorted = sorted(freqs_sorted, key=lambda item: item[1])
        
    return freqs_sorted[0][0]
    #return root node (so we can use this generating coding....)

#input
message = 'AAABBBBBBEEEDABEEDCC'

#count the letters
#use Counter, then convert to dictionary
freqs = dict(Counter(message)) #{'A': 4, 'B': 7, 'E': 5, 'D': 2, 'C': 2}
# print(freqs['A'])  #4

#sort them from smallest to biggest
#{'C': 2, 'D': 2, 'A': 4, 'E': 5, 'A': 7}
freqs_sorted = sorted(freqs.items(), key=lambda item: item[1])

#make the tree by combining the smallest one, and delete those guys
root = make_the_tree(freqs_sorted)

#get the code
huffman_code = get_code(root)

#print the code
print(huffman_code)
#{'A': '00'; 'B': '11'; 'C': '011'; 'D': '010'; 'E': '10'}

encd_msg = '0001001001101010' # ADDCDE
#task1: decode the encoded message to the original message
original_message = decode(encd_msg, huffman_code)
print(original_message)

# #task2: calculate the total cost --> message + table
print('Message:', message)
calculateTotalCost(dict(freqs_sorted), huffman_code)

# for cal
```
