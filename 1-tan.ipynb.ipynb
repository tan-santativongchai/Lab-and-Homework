# INSERTION SORT
## Pseudo Code (Ascent)
>for j = 2 to A.length<br>
>>    key = A[j]<br>
>>    //Insert A[j] in to the sorted sequence A[1 . . j-1]<br>
>>    i = j - 1<br>
>>    while i > 0 and A[i] > key:<br>
>>>        A[i+1] = A[i]<br>
>>>        i = i-1<br>
>>    A[i+1] = key <br>

For Insertion Sort I use the pseudo code directly.

```python
A = np.array(tuple(arr_test))
for j in range(len(A))[1:]:
    k = A[j]
    i = j - 1
    while i >= 0 and A[i] > k:
        A[i + 1] = A[i]
        i = i-1
    A[i + 1] = k

#print the original array compare when use insertion sort.
print("ORIGINAL ORDER")
print(arr_test)
print("INSERTION SORT (Ascent)")
print(A)
#Make the array to check if my merge sort is correctly working or not
Test_i_as = np.array(tuple(A))
```
